Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np

mu = 3
sigma = 0.5
n=1000
vals = np.random.normal(loc=mu, scale=sigma, size=n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np

# Parámetros de la distribución log-normal
mean_log = 0   # Media del logaritmo natural
sigma_log = 0.5  # Desviación estándar del logaritmo natural
n = 1000

vals =  np.random.lognormal(mean_log, sigma_log, size=n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad normal estándar.

```python
from scipy.stats import lognorm
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución log-normal
mu = 0     # Media del logaritmo natural
sigma = 1  # Desviación estándar del logaritmo natural
x = np.linspace(0, 6, 1000)  # Rango de valores para la x

# Calcular la función de densidad de la distribución log-normal
lognormal_pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))

# Graficar la función de densidad de la distribución log-normal
plt.plot(x, lognormal_pdf, label=f'Log-Normal $\mu={mu}$, $\sigma={sigma}$')
plt.title('Función de densidad distribución log-normal')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de una distribución log-normal con mu=0 y sigma=1
mu_lognormal = 0
sigma_lognormal = 1

# Crear un QQ-plot para comparar con la distribución log-normal
fig = sm.qqplot(data=vals, dist=stats.lognorm, line="45",fit = False, distargs= (0.5,), loc=mu_lognormal, scale=sigma_lognormal)

# Configuraciones de la gráfica
plt.title(f'QQ-plot para Log-Normal ($\mu=0$, $\sigma=1$)')
plt.xlabel('Cuantiles teóricos')
plt.ylabel('Cuantiles de los datos')

# Mostrar la gráfica
plt.show()
```
